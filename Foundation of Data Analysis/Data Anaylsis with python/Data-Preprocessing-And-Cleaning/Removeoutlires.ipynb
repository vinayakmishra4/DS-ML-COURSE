```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the dataset
df=pd.read_csv('googleplaystore.csv')
```

```python
df.drop_duplicates(inplace=True)
```

```python
plt.subplot(1, 2, 1)
sns.distplot(df['Rating'])
```

```python
mean=df['Rating'].mean()
std=df['Rating'].mean()
```

```python
print(mean , std)
```

```python
Highset_allowed=mean + 3*std
lowset_allowed=mean-3*std
```

```python
print(Highset_allowed,lowset_allowed)
```

# Triming

```python
df[(df['Rating'] < Highset_allowed) | (df['Rating'] > lowset_allowed)]
```

```python
df[(df['Rating'] > lowset_allowed) & (df['Rating'] < Highset_allowed)]
```

```python
# Load your dataframe (replace with your real data)
df = pd.read_csv("googleplaystore.csv")

# Calculate Z-score for the Rating column
df['Rating_zscore'] = (df['Rating'] - df['Rating'].mean()) / df['Rating'].std()

# Filter out outliers
df_no_outliers = df[df['Rating_zscore'].abs() <= 3].copy()

# Optionally drop the z-score helper column
df_no_outliers.drop(columns=['Rating_zscore'], inplace=True)

df_no_outliers.head()
```
