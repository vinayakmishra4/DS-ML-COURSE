## CREATING PANDAS SERIES

```python
# create the list and element in list
list=[1,2,3,4]
# now make the list into pandas series by the Series function
pd1=pd.Series(list)
print(pd1)
```

```python
# create the array using numpy 
arr=np.array([1,2,3,4,5])
# now make the array into pandas series by the Series function
pd2=pd.Series(arr)
print(pd1)
```

```python
# Series index in the Series
pd3=pd.Series(data=list,index=['Columuns1','Columuns2','Columuns3','Columuns4'])
print(pd3)
```

```python
# creating the dictionary
step={'day1':6000,'day2':6600,'day3':6660,'day4':6666}
# now create the convert the dictionary to pandas Series using Series function 
pd4=pd.Series(step)
print(pd4)
```
