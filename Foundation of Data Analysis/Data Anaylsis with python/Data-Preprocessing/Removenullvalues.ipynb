```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

# LOAD DATASET

```python
df=pd.read_csv('googleplaystore.csv')
```

```python
data=pd.DataFrame(df)
```

# Check the Null Values

```python
# isnull() function will tell whether null values is there or not
# sum() this function add all the null values and print 
data.isnull().sum()
```

# REMOVE NULL VALUES

```python
 data.dropna()
# this will remove all the null values form the dataset
```

```python
# this method is not recommended for the large datasets beacasue it losses the values of the dataset
# 15 % loss in dataset
```

# ALTER THE DATASET

## Imputer method on numerical data

```python
df.dtypes
```

```python
# this method will replace all the null values with the mean
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
```

```python
# it will the mean calulate the mean 
imputer.fit(df[['Rating']])
```

```python
# it will replace the null values with mean values
df['Rating'] = imputer.transform(df[['Rating']])
```

```python
df.isnull().sum()
```

## Imputer method on the catetorycal data

```python
df['Type'] = SimpleImputer(strategy='most_frequent').fit_transform(df[['Type']]).ravel()
```

```python
df.isnull().sum()
```
