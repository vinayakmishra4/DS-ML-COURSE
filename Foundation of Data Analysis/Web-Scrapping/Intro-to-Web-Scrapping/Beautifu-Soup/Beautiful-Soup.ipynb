# 🌐 Beautiful Soup

## Step 1: Import libraries

```python
# importing necessary libraries to scrape data from the web and save it in a CSV file
import requests
from bs4 import BeautifulSoup
import csv
```

## Step 2: Get the webpage

```python
# URL of the website to scrape data
URL = "https://www.passiton.com/inspirational-quotes"
# sending a GET request to the website and parsing the HTML content using BeautifulSoup
response = requests.get(URL)
# parsing the HTML content using BeautifulSoup
soup = BeautifulSoup(response.text, "html.parser")
# checking if the request was successful by printing the status code
print(response.status_code)  # should be 200
```

# Step 3:Searching and navigating through the parse tree

```python
# extracting the quotes from the HTML content and storing them in a list of dictionaries
quotes = []
# finding the div with id 'all_quotes' and then finding all the divs with the specified class to extract the quote information
table = soup.find('div', id='all_quotes')
# checking if the table was found before trying to extract the quotes
if table:
    rows = table.find_all('div',class_='col-6 col-lg-3 text-center margin-30px-bottom sm-margin-30px-top')
    # iterating through each row and extracting the quote information such as theme, url, image source, and lines of the quote
    for row in rows:
        quote = {}
        quote['theme'] = row.find('h5').text.strip() if row.find('h5') else ""
        quote['url'] = row.find('a')['href'] if row.find('a') else ""
        quote['img'] = row.find('img')['src'] if row.find('img') else ""
        quote['lines'] = row.find('img')['alt'].split(" #")[0] if row.find('img') else ""

        quotes.append(quote)
```

## Step 5: Save data to CSV

```python
# saving the extracted quotes to a CSV file named "quotes.csv" with the specified fieldnames
with open("quotes.csv", "w", newline="", encoding="utf-8") as f:
    # creating a DictWriter object and writing the header and rows to the CSV file
    writer = csv.DictWriter(f, fieldnames=['theme','url','img','lines'])
    # writing the header and rows to the CSV file
    writer.writeheader()
    # writing the rows to the CSV file
    writer.writerows(quotes)

# printing a message indicating that the scraping is completed and the total number of quotes extracted
print("Scraping completed. Total quotes:", len(quotes))
```
