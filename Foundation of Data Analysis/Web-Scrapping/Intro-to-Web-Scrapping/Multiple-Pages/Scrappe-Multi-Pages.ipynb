# Scrapping Multiple Pages

## Step 1: Fetch One Page First

```python
import requests
from bs4 import BeautifulSoup
# suppress warnings about unverified HTTPS requests
import urllib3

# Disable warnings about unverified HTTPS requests
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

url = "https://example.com/page=1"
headers = {"User-Agent": "Mozilla/5.0"}

# add the verify=False parameter to the get request to ignore SSL certificate verification
response = requests.get(url, headers=headers, verify=False)

soup = BeautifulSoup(response.text, "html.parser")
print(soup.title.text)
```

## Step 2: Extract Data from One Page

```python
items = soup.find_all("h2")  # change this!

# print the text of each item
for item in items:
    print(item.text.strip())
```

# Step 3: Loop Through Multiple Pages

```python
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)
headers ={"User-Agent": "Mozilla/5.0"}
base_url = "https://books.toscrape.com/catalogue/page-{}.html"

# loop through pages 1 to 5
# Note: Adjust the range as needed based on the actual number of pages available
for page in range(1, 6):
    url = base_url.format(page)
    print(f"Scraping page {page}...")
    response = requests.get(url, headers=headers, verify=False)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")
        items = soup.find_all("h2")  # change selector
        for item in items:
            print(item.text.strip())
    else:
        print("Failed:", url)
```

## Step 7: Store Data 

```python
# Now let's store the data in a list instead of printing it
data = []

# loop through pages 1 to 5 and store the data
for page in range(1, 6):
    url = base_url.format(page)
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")
    items = soup.find_all("h2")
    for item in items:
        # Instead of printing, we append the text to our data list
        data.append(item.text.strip())

# Finally, we can print the collected data
print(data)
```
