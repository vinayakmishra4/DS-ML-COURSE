# Scrapping Quotes

## Step 1: Importing Libraries and Connecting to the Website

```python
# import the necessary libraries and make a request to the webpage and then parse the webpage using BeautifulSoup
import requests
from bs4 import BeautifulSoup
import pandas as pd
link = 'https://quotes.toscrape.com/'
res = requests.get(link)
soup = BeautifulSoup(res.text, 'html.parser')
```

## Step 2: Scarping Quotes

```python
# Extracting the quotes from the webpage
quotes = []
for quote in soup.find_all('span', class_='text'):
    quotes.append(quote.text[1:-1])
    print(quote.text[1:-1])
```

## Step 3: Scarping Author Names

```python
# Extracting the Authors of the quotes
authors = []
for quote in soup.find_all('small', class_='author'):
    authors.append(quote.text)
    print(quote.text)
```

##  Step 4: Scrapping Proflie Link


```python
# Extracting the profile links of the authors
profile_link = []

for quote in soup.find_all('div', class_='quote'):
    link = quote.find('a').get('href')
    profile_link.append(link)
    print(link)
```

##  Step 4: Scrapping Tags of the quotes

```python
# Extracting the tags of the quotes
tags=[]
for quote in soup.find_all('div', class_='quote'):
    tag = quote.find_all('a', class_='tag')
    tags.append([t.text for t in tag])
    print([t.text for t in tag])
```

## Step 5:Store the Data in the list

```python
# Sore the data in a list of dictionaries
Data=[]
for i in range(len(quotes)):
    Data.append({
        'Quote': quotes[i],
        'Author': authors[i],
        'Profile Link': profile_link[i],
        'Tags': tags[i]
    })
```

## Step 6: Convert into Dataframe

```python
# Creating a dataframe from the data
df=pd.DataFrame(Data, columns=['Quote', 'Author', 'Profile Link', 'Tags'])
```

## Step 8: Scrapping Multiple Pages

```python
# Now we will extract the data from multiple pages of the website
# this library is used to show the progress of the loop
from tqdm import tqdm
Multiple_Pages = []
for page in tqdm(range(1, 11)):
    link = ('http://quotes.toscrape.com/page/' + str(page))
    res = requests.get(link)
    soup = BeautifulSoup(res.text, 'html.parser')

    for sp in soup.find_all('div', class_='quote'):
        quote = sp.find('span', class_='text').text[1:-1]
        authors = sp.find('small', class_='author').text
        details = sp.find('a').get('href')
        tags = []

        for tag in sp.find_all('a', class_='tag'):
            tags.append(tag.text)

        # Joining the tags in a single string   
        tags = ','.join(tags)
        Multiple_Pages.append([quote, authors, details, tags])
```

```python
Multiple_Pages_df = pd.DataFrame(data=Multiple_Pages)
Multiple_Pages_df = Multiple_Pages_df.rename(
    columns={0: 'Quote', 1: 'Author', 2: 'Author_id', 3: 'Tags'})
Multiple_Pages_df['Author_Link'] = 'http://quotes.toscrape.com/' + \
    Multiple_Pages_df['Author_id']
Multiple_Pages_df.head()

```

## Step 7: Save to CSV

```python
# Saving the data in an excel file
with pd.ExcelWriter('Quotes.xlsx') as writer:
    Multiple_Pages_df.to_excel(writer, sheet_name='Quotes', index=False)
```
